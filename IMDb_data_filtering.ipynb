```python
###------------Importing Necassary Libraries-------------------####

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

###------------------Data Filtering for the 'titles' File -----------------###

## Read the file ##

df_title = pd.read_csv("title.basics.tsv",sep='\t', low_memory=False)

## Drop unnecessary Columns ##

df_title.drop(['isAdult','endYear'], axis=1, inplace=True)

## What is the types of data present ##

df_title.dtypes

## Looking into movies only ##

df_title["startYear"] = pd.to_numeric(df_title["startYear"], errors='coerce')

df_title.head()

df_title["titleType"].unique()

df_movies = df_title[df_title["titleType"] == "movie"]

df_movies.head()

## Puicking movies after 2017 ##


df_movies = df_movies[df_movies["startYear"] == 2017]

df_movies.shape[0]

df_movies["startYear"].unique

df_movies['genres'].unique()
```

```python
df_ratings = pd.read_csv("title.ratings.tsv",sep='\t', low_memory=False)
df_ratings.head()
```

```python
## Merging the df_ratings dataframe with the df_movies dataframe ##

df_mratings = pd.merge(df_movies,df_ratings, on='tconst',how='outer')
df_mratings.head()
```

```python
df_crew = pd.read_csv("title.crew.tsv", sep='\t', low_memory=False)
df_crew.head()
```

```python
## Doing an over all merge ##

df_final = pd.merge(df_mratings,df_crew, on='tconst',how='outer')
df_final.head()


df_final.drop_duplicates('tconst', inplace = True) 
df_final.shape[0]
```
